# Bonus: Which pair of genres is the most popular?

```python
def convert_nums(x, genre):
    if x is np.nan:
        return np.nan
    elif x == 1:
        return genre
    else:
        return np.nan
```

```python
for genre in genres.values():
    merged_df_pl_int[genre] = list(map(lambda x: convert_nums(x, genre), merged_df_pl_int[genre]))
```
